```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 120, 300)

Grain = 0.3 + 0.6 * (1 / (1 + np.exp(-0.05*(t-60))))   # sigmoid rise
Anchory = 0.9 * np.exp(-0.03*t)                        # exponential decay
CAP_adj = 0.5 + 0.2 * (1 - np.exp(-0.02*t))             # gradual curve up

plt.plot(t, Grain, label='Grain(t)')
plt.plot(t, Anchory, label='Anchory(t)')
plt.plot(t, CAP_adj, label='CAP_adj(t)')

plt.xlabel('Time (seconds)')
plt.ylabel('Value')
plt.title('Coupled Dynamics of Grain(t), Anchory(t), and CAP_adj(t)')
plt.legend()
plt.grid(True)
plt.show()
```
